## Documentation 

Creating a basic Image Manipulation Tool using Numpy. 
Matplotlib has also been used to plot the image.
Skimage has been used to include the image in the python file.

## Code Snippet

```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
import time

def Load(image):                                             #load image
    plt.imshow(image)
    plt.show()
    print("The image has been successfully loaded")
    
def Hf(image):
    img_hf=image[:, ::-1]                                    #horizontal flip
    plt.imshow(img_hf)
    plt.show()
    return img_hf
    
def Vf(image):
    img_vf=np.flipud(image)                                  #vertical flip
    plt.imshow(img_vf)
    plt.show()
    return img_vf
    
def Ic(image):                                               #invert colors
    img_ic=image
    img_ic=255-img_ic
    plt.imshow(img_ic)
    plt.show()
    return img_ic
    
def Gray(image):
    img_grey=np.dot(image[...,:3], [0.299, 0.587, 0.144])    #grayscale conversion
    plt.imshow(img_grey,cmap='gray')
    plt.show()
    return img_grey
    
def B_W(image):
    img_bw=np.dot(image[...,:3], [0.299, 0.587, 0.144])     #black and white conversion
    plt.imshow(img_bw,cmap='gray')
    plt.show()
    return img_bw
    
def Save(image1):                                           #save image      
    plt.imshow(image1)
    plt.savefig("image1.png")
    print("Image was succesfully saved")

    
image=io.imread("582092-minion-desktop.png")      #originalimage
image1=io.imread("582092-minion-desktop.png")     #for editing image
while True:
        print("\n Main Menu")
        print("1. Load Image")
        print("2. Edit Iamge")
        print("3. Save Image")
        print("4. Quit")
        ch=int(input("Enter Your Choice: "))
        
          #reading the image
        
        
        if ch==1:
            Load(image)
            time.sleep(1)
            
        elif ch==2:
            print("\n What do you want to do with the image:??")
            print("1. Horizontal Flip")
            print("2. Vertical Flip")
            print("3. Invert Colors")
            print("4. Convert to Grayscale")
            print("5. Convert to B&W")
            ch2=int(input("Enter the option: "))
            
            if ch2==1:
                image1=Hf(image)
                time.sleep(1)
            elif ch2==2:
                image1=Vf(image)
                time.sleep(1)
            elif ch2==3:
                image1=Ic(image)
                time.sleep(1)
            elif ch2==4:
                image1=Gray(image)
                time.sleep(1)
            elif ch2==5:
                image1=B_W(image)
                time.sleep(1)
            else:
                print("Invalid choice")
                
        elif ch==3:
            Save(image1)
            time.sleep(1)
        
        elif ch==4:
            print("The program is succesfully executed")
            break
            
        else:
            print("Invalid Choice")
```
